```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Copy of  Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Copy of  Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel("/content/Copy of  Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_chatgpt = pd.read_excel("/content/Copy of  Imersão Python - Tabela de ações.xlsx", sheet_name="ChatGPT")
df_chatgpt.head(10)
```

```python
df_principal = df_principal [['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Último (R$)': 'valor_final', 'Var. Dia (%)': 'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] +1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_chatgpt
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade (anos)'].apply(lambda x: 'Mais que 100 anos' if x > 100 else ('Menos que 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
# Calculando o maior valor da coluna Variacao_rs
maior_valor = df_principal['Variacao_rs'].max()

# Calculando o menor valor da coluna Variacao_rs
menor_valor = df_principal['Variacao_rs'].min()

# Calculando a média dos valores da coluna Variacao_rs
media_total = df_principal['Variacao_rs'].mean()

# Calculando a média dos valores da coluna Variacao_rs onde Resultado é 'Subiu'
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média dos valores da coluna Variacao_rs onde Resultado é 'Desceu'
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Exibindo os resultados
print("Maior\t", f"R$ {maior_valor:,.2f}")
print("Menos\t", f"R$ {menor_valor:,.2f}")
print("Media\t", f"R$ {media_total:,.2f}")
print("Media de quem subiu\t", f"R$ {media_subiu:,.2f}")
print("Media de quem desceu\t", f"R$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variacao Reais por Resultado')
fig.show()
```
